```python
!pip install catboost
import pandas as pd

car_file_path = '/content/car_prediction_data.csv'
car_data = pd.read_csv(car_file_path)
car_data = car_data.dropna(axis=0)

car_data = pd.get_dummies(
    car_data,
    columns=['Fuel_Type', 'Seller_Type', 'Transmission'],
    drop_first=True
)

#target
y = car_data['Selling_Price']

car_features = [
    'Year',
    'Present_Price',
    'Kms_Driven',
    'Owner',
    'Fuel_Type_Diesel',
    'Fuel_Type_Petrol',
    'Seller_Type_Individual',
    'Transmission_Manual'
]

X = car_data[car_features]

from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(
    X,
    y,
    train_size=0.8,
    random_state=1
)

from catboost import CatBoostRegressor

model = CatBoostRegressor(
    iterations=500,
    learning_rate=0.05,
    depth=6,
    loss_function='RMSE',
    random_seed=1,
    verbose=0
)

model.fit(X_train, y_train)

predictions = model.predict(X_valid)

from sklearn.metrics import mean_absolute_error, r2_score

mae = mean_absolute_error(y_valid, predictions)
r2 = r2_score(y_valid, predictions)

print("Mean Absolute Error:", mae)
print("R² Score:", r2)
comparison = pd.DataFrame({
    'Actual Price': y_valid.values,
    'Predicted Price': predictions
})

print(comparison.head(10))
```
