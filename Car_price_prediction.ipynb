```python
import pandas as pd
car_file_path = '/content/car_prediction_data.csv'
car_data = pd.read_csv(car_file_path)
car_data = car_data.dropna(axis=0)

car_data = pd.get_dummies(
    car_data,
    columns=['Fuel_Type', 'Seller_Type', 'Transmission'],
    drop_first=True
)

#target
y = car_data.Selling_Price

car_features = [
    'Year',
    'Present_Price',
    'Kms_Driven',
    'Owner',
    'Fuel_Type_Diesel',
    'Fuel_Type_Petrol',
    'Seller_Type_Individual',
    'Transmission_Manual'
]
X = car_data[car_features]
z = X.head()

from sklearn.tree import DecisionTreeRegressor

car_model = DecisionTreeRegressor(random_state=1)

car_model.fit(X, y)

print("Making predictions for the following 5 cars:")
print(z)

print("The predictions are")
print(car_model.predict(z))
```
